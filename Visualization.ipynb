```python
t_pre = 4
h = t_pre-1  # which time
prop = 0.5

```

```python
OD_2019 = np.load('/Volumes/GoogleDrive/My Drive/Graduate/SP22 CE 299/data/BART/hour data/3d.npy')

plt.figure(figsize=[20,5])
plt.plot(OD_2019[org,dest,360+h:716+h], '.-')  
data = np.load('./res/%.1f/%s_%i.npy'%(0.5, 'labels', 2019)).T.reshape(4, 48, 48, -1)  # start from 362
plt.plot(data[h, org, dest, :], '.-')
```

```python

year = 2019

model_lst = ['labels', 'conv', 'convgraph', 'lr']
legend_lst = ['Ground Truth', 'CNN', 'GCN', 'LR']
color_lst = ['blue', 'orange', 'red', 'green']
lw_lst = np.array([1., .5, .5, .5])*2
alpha_lst = [1, .7, .9, .7]

plt.figure(figsize=[15,5], dpi=225)

for i in range(4):
    model = model_lst[i]
    color = color_lst[i]
    path = './res/%.1f/%s_%i.npy'%(prop, model, year)
    data = np.load(path)
    data = data.T.reshape(4, 48, 48, -1)
    plt.plot(data[h, org, dest, :], '.-', lw=lw_lst[i], alpha = alpha_lst[i], color=color, label=legend_lst[i])

plt.legend()
plt.grid(ls='--')

x_coor = []
x_coor_ticks = []
for i in range(16):
    x_coor.append(360+i*24-360-h)  # 360 is 15th day of Jan, data starts from 362
    x_coor_ticks.append('%i-1-%i  00:00'%(year, 16+i))
plt.xticks(x_coor, x_coor_ticks, rotation=15)
plt.xlabel('Date')
plt.ylabel('Trip counts')
```
