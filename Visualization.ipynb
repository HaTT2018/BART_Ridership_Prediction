```python
prop = 0.5
year = 2019
t_pre = 4

model_lst = ['labels', 'conv', 'convgraph', 'lr']
color_lst = ['blue', 'orange', 'red', 'green']
lw_lst = np.array([1., .5, .5, .5])*2
alpha_lst = [1, .7, .9, .7]

plt.figure(figsize=[15,5], dpi=225)
h = t_pre-2  # which time
org = 17
dest = 20
for i in range(4):
    model = model_lst[i]
    color = color_lst[i]
    path = './res/%.1f/%s_%i.npy'%(prop, model, year)
    data = np.load(path)
    data = data.T.reshape(4, 48, 48, -1)
    plt.plot(data[h, org, dest, :], '.-', lw=lw_lst[i], alpha = alpha_lst[i], color=color, label=model)

plt.legend()
plt.grid(ls='--')

x_coor = []
x_coor_ticks = []
for i in range(15):
    x_coor.append(360+i*24-362)  # 360 is 15th day of Jan, data starts from 362
    x_coor_ticks.append('%i-1-%i  00:00'%(year, 16+i))
plt.xticks(x_coor, x_coor_ticks, rotation=30)
```
